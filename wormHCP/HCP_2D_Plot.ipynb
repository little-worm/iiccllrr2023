```python
import matplotlib.pyplot as plt
import numpy as np
import os
from nilearn import plotting 
import nibabel as nibaptitu
#pwd = os.path.dirname(__file__)
pwd =  os.path.dirname(os.path.realpath('__file__'))
pwd = pwd + '/'
father_path=os.path.abspath(os.path.dirname(pwd)+os.path.sep+".")
father_path = father_path + '/' 
grader_father_path=os.path.abspath(os.path.dirname(pwd)+os.path.sep+"..")
grader_father_path = grader_father_path + '/'
print("pwd                = ",pwd)
print("father_path        = ",father_path)
print("grader_father_path = ",grader_father_path)

 
import nibabel as nib
 
def read_nii_file1(nii_path):
    '''
    根据nii文件路径读取nii图像
    '''
    nii_image=nib.load(nii_path)
    return nii_image
 
def nii_one_slice1(image):
    '''
    显示nii image中的其中1张slice
    '''
    image_arr=image.get_fdata()[:,:,:,5]
    print(type(image_arr))
    print(image_arr.shape)
    #注意：nibabel读出的image的data的数组顺序为：Width，Height，Channel
    #将2d数组转置，让plt正常显示
    for img_num in range(109):
        image_2d=image_arr[:,img_num,:].transpose((1,0))
        print(image_2d.shape)
        plt.imshow(image_2d, cmap = 'gray')
        #plt.savefig(imgPath+'/test.png')
        plt.show()
        print("img_num = ",img_num)

imgPath = pwd
plt.savefig(imgPath+'/test.png')
example_filename = pwd + 'fourDimBrains/996782_EMOTION.nii.gz'  
example_filename = pwd + 'fourDimBrains/996782_WM.nii.gz'  
example_filename = pwd + 'fourDimBrains/996782_SOCIAL.nii.gz'  
nii_image1=read_nii_file1(example_filename)
print(nii_image1.shape)
nii_one_slice1(nii_image1)









# example_filename = '/home/worm/wormCode/wormPyCode20210901/HCP/niipre_master/996782_EMOTION.nii.gz' 

#plt.savefig(imgPath+'/test.png')
# /home/worm/anaconda3/envs/worm/bin/python /home/worm/wormCode/wormPyCode20210901/HCP/niipre_master/hcpread.py 






```
